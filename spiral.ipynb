```python
arr1 = []

# Open the CSV file and read it
with open("fort.1", "r") as file:
    # Create a CSV reader, specifying that the delimiter is a comma
    csv_reader = csv.reader(file, delimiter=',')
    
    # Iterate through the rows in the CSV file
    for row in csv_reader:
        # Remove the empty string at the beginning of each row
        row_data = row[1:]
        
        # Convert the row values to integers
        row_data = [int(value) for value in row_data]
        
        # Append the row data to the arr list
        arr1.append(row_data)

# Now, arr contains your grid as a 2D list
arr1 = np.array(arr1)
print("Biggest matrix inversion in DI=",arr1.size)
```
